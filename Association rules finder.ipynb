```python
#Packages

import pandas as pd 

import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder

from mlxtend.frequent_patterns import apriori
```

```python
df = pd.read_csv("Products visits.csv")
df = df[['User','ProductId']]
df
```

```python
df = df.groupby('User')['ProductId'].apply(list).reset_index()
df.drop(['User'], axis=1, inplace=True) 


column_list = df['ProductId'].tolist()


from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_data = te.fit(column_list).transform(column_list)
df = pd.DataFrame(te_data,columns=te.columns_).astype(int)
df
```

```python
# Apriori
freq_items = apriori(df, min_support = 0.1, use_colnames = True, verbose = 1)
freq_items.sort_values("support", ascending = False).head(20)
freq_items.to_json('ArticlesSimilaires.json', orient='records')
```
